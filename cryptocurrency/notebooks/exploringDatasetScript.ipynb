```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url="../data/processed/dataset_output.csv"
parse_dates=["Date"]
ds = pd.read_csv(url, parse_dates=parse_dates)
```

```python
ds.shape
```

```python
ds.tail(10)
```

```python
ds.describe(include="all")
```

```python
dsb = ds[pd.notnull(ds["Price_bitcoin"])]
dsb = dsb[["Date","Price_bitcoin","MarketCap_bitcoin","BlockSize_bitcoin","HashRate_bitcoin","Difficulty_bitcoin","Address_bitcoin"]]
```

```python
dsb.sample(10)
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
for col in list(dsb)[1:]:
    plt.plot(dsb["Date"], dsb[col], label=col)
plt.xlabel("Date")
plt.ylabel("Cryptocurrency")
plt.legend()
```

```python
# Subsert for interesting Data value
ds_subset = dsb[dsb["Date"] > "2015-10-01"]
```

```python
fig, ax = plt.subplots()
ds_subset.Price_bitcoin.plot(ax=ax, style='b')
ax.set_ylabel('Price')

# same ax as above since it's automatically added on the right
ds_subset.BlockSize_bitcoin.plot(ax=ax,secondary_y=True, style='g')
ax.right_ax.set_ylabel('BlockSize')

ax3 = ax.twinx()
rspine = ax3.spines['right']
rspine.set_position(('axes', 1.1))
ax3.set_frame_on(True)
ax3.patch.set_visible(False)
fig.subplots_adjust(right=0.7)
ds_subset.HashRate_bitcoin.plot(ax=ax3, style='r')
ax3.set_ylabel('HashRate')

ax4 = ax.twinx()
rspine = ax4.spines['right']
rspine.set_position(('axes', 1.2))
ax4.set_frame_on(True)
ax4.patch.set_visible(False)
fig.subplots_adjust(right=0.7)
ds_subset.Difficulty_bitcoin.plot(ax=ax4, style='y')
ax4.set_ylabel('Difficulty')

ax5 = ax.twinx()
rspine = ax5.spines['right']
rspine.set_position(('axes', 1.3))
ax5.set_frame_on(True)
ax5.patch.set_visible(False)
fig.subplots_adjust(right=0.7)
ds_subset.Address_bitcoin.plot(ax=ax5,style='#88ee88')
ax5.set_ylabel('Address')

ax3.legend([ax.get_lines()[0],ax.right_ax.get_lines()[0],ax3.get_lines()[0],ax4.get_lines()[0],ax5.get_lines()[0]], 
           ['Price','BlockSize','HashRate','Difficulty','Address'], bbox_to_anchor=(0.15, 0.95))



#ax3.legend([ax.get_lines()[0], ax.right_ax.get_lines()[0], ax3.get_lines()[0]],\
#           ['Price','BlockSize','MarketCap'], bbox_to_anchor=(1.5, 0.5))
```
