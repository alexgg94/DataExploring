```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url="../data/processed/sorted_output.csv"
parse_dates=["Date"]
ds = pd.read_csv(url, parse_dates=parse_dates)
```

```python
ds.shape
```

```python
ds.head(10)
```

```python
ds.describe(include="all")
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
for cryptocurrency in list(ds)[1:]:
    plt.plot(ds["Date"], ds[cryptocurrency], label=cryptocurrency)
plt.xlabel("Date")
plt.ylabel("Cryptocurrency")
plt.legend()
```

```python
ds_subset = ds[ds["Date"] > "2017-01-01"]
```

```python
"""
Subset for interesting Date value.
Removed bitcoin for the better visualization of the remaining cryptocurrencies.
"""

for cryptocurrency in list(ds_subset)[2:]:
    plt.plot(ds_subset["Date"], ds_subset[cryptocurrency], label=cryptocurrency)
plt.xlabel("Date")
plt.ylabel("Cryptocurrency")
plt.legend()
```
